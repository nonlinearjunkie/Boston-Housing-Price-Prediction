```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

import sklearn

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

```python
print(boston_dataset.DESCR)
```

```python
boston_df=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["Price($1000)"]=boston_dataset.target
```

```python
boston_df.head()
```

```python
boston_df["Price($1000)"].describe()
```

```python
sns.set(rc={'figure.figsize':(13,11)})
sns.distplot(boston_df["Price($1000)"], bins=30)
plt.show()
```

```python
correlation_matrix = boston_df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
X=boston_df.drop("Price($1000)", axis = 1)
Y=boston_df["Price($1000)"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
y_predict = lin_model.predict(X_test)
print("Root mean squared error: {}".format(np.sqrt(mean_squared_error(Y_test, y_predict))))
```
